```python
import numpy as np
import matplotlib.pyplot as plt
embedding = np.load("/data1/ReactionT5_neword/task_forward/output_ord/embedding_mean.npy")
print(embedding.shape)
```

```python
# apply PCA and plot
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
embedding_pca = pca.fit_transform(embedding)
print(embedding_pca.shape)
```

```python
fig, ax = plt.subplots()
ax.scatter(embedding_pca[:, 0], embedding_pca[:, 1], alpha=0.1, s=0.5)
```

```python
embedding_USPTO = np.load("/data1/ReactionT5_neword/task_forward/output_uspto/embedding_mean.npy")
print(embedding_USPTO.shape)
```

```python
embedding_USPTO_pca = pca.transform(embedding_USPTO)
print(embedding_USPTO_pca.shape)
```

```python
fig, ax = plt.subplots()
ax.scatter(embedding_pca[:, 0], embedding_pca[:, 1], alpha=0.1, s=0.5)
ax.scatter(embedding_USPTO_pca[:, 0], embedding_USPTO_pca[:, 1], alpha=0.1, s=0.5)
```
