```python
import glob
import os
import json
train_losses = [[], [], [], [],[], [], [], []]
eval_losses = [[], [], [], [],[], [], [], []]
steps = [[], [], [], [],[], [], [], []]
eval_steps = [[], [], [], [], [], [], [], []]
for ith, dir_name in enumerate(['lr001_wd0001', 'lr0001_wd0001', 'lr00001_wd0001', 'lr000001_wd0001', 'lr0001_wd01', 'lr0001_wd001', 'lr0001_wd00001', 'lr0001_wd000001']):
    dir = sorted(glob.glob(os.path.join(dir_name, 'output/checkpoint*')))[-1]
    with open(os.path.join(dir, 'trainer_state.json')) as f:
        # read file as a json object
        data = json.load(f)["log_history"]
        for i in range(len(data)):
            if 'eval_loss' in data[i]:
                eval_losses[ith].append(data[i]["eval_loss"])
                eval_steps[ith].append(data[i]["step"])
            else:
                train_losses[ith].append(data[i]["loss"])
                steps[ith].append(data[i]["step"])
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 10))
plt.subplot(2, 1, 1)
plt.title('Train Loss')
plt.xlabel('Step')
plt.ylabel('Loss')
plt.plot(steps[0], train_losses[0], label='lr=0.01, wd=0.001', alpha=0.5)
plt.plot(steps[1], train_losses[1], label='lr=0.001, wd=0.001', alpha=0.5)
plt.plot(steps[2], train_losses[2], label='lr=0.0001, wd=0.001', alpha=0.5)
plt.plot(steps[3], train_losses[3], label='lr=0.00001, wd=0.001', alpha=0.5)
plt.plot(steps[4], train_losses[4], label='lr=0.001, wd=0.1', alpha=0.5)
plt.plot(steps[5], train_losses[5], label='lr=0.001, wd=0.01', alpha=0.5)
plt.plot(steps[6], train_losses[6], label='lr=0.001, wd=0.0001', alpha=0.5)
plt.plot(steps[7], train_losses[7], label='lr=0.001, wd=0.00001', alpha=0.5)
# plt.ylim(0, 10)
plt.legend()
plt.subplot(2, 1, 2)
plt.title('Eval Loss')
plt.xlabel('Step')
plt.ylabel('Loss')
plt.plot(eval_steps[0], eval_losses[0], label='lr=0.01, wd=0.001', alpha=0.5)
plt.plot(eval_steps[1], eval_losses[1], label='lr=0.001, wd=0.001', alpha=0.5)
plt.plot(eval_steps[2], eval_losses[2], label='lr=0.0001, wd=0.001', alpha=0.5)
plt.plot(eval_steps[3], eval_losses[3], label='lr=0.00001, wd=0.001', alpha=0.5)
plt.plot(eval_steps[4], eval_losses[4], label='lr=0.001, wd=0.1', alpha=0.5)
plt.plot(eval_steps[5], eval_losses[5], label='lr=0.001, wd=0.01', alpha=0.5)
plt.plot(eval_steps[6], eval_losses[6], label='lr=0.001, wd=0.0001', alpha=0.5)
plt.plot(eval_steps[7], eval_losses[7], label='lr=0.001, wd=0.00001', alpha=0.5)
plt.legend()
plt.savefig('loss.png')
plt.show()
```
