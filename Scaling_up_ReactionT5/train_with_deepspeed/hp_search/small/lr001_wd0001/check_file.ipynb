```python
import glob
import pandas
data_files = glob.glob('/media/sagawa//7182ee6c-8215-4bea-a609-999c7c2c02cf/preprocessed_ZINC22/*.parquet')
# get file sizes without loading the data
filesizes = {}
for data_file in data_files:
    with open(data_file, 'rb') as f:
        f.seek(0, 2)
        filesizes[data_file] = f.tell()

data_files = [k for k, v in sorted(filesizes.items(), key=lambda item: item[1])]
print(data_files)
for data_file in data_files:
    df = pandas.read_parquet(data_file)
    df['len'] = df['input_ids'].apply(lambda x: len(x))
    print(data_file)
    print(df['len'].describe())
```
